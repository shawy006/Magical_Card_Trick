```python
def display(pool_A,pool_B,pool_C):
    print()
    print()
    print("Does your number lies in pool_A,pool_B,pool_C?")
    print()
    print("Pool_A  -->" , random.sample(pool_A,len(pool_A)),"        ~~~~~~~   <<<< If there press 'a' ")
    print()
    print()
    print("Pool_B  -->" , random.sample(pool_B,len(pool_B)),"        ~~~~~~~   <<<< If there press 'b' ")
    print()
    print()
    print("Pool_C  -->" , random.sample(pool_C,len(pool_C)),"        ~~~~~~~   <<<< If there press 'c' ")

def InWhichPool(pool_A,pool_B,pool_C,user):
    if user=='a':
        A=pool_C[::-1]+pool_A[::-1]+pool_B[::-1]
        return A
    if user=='b':
        B=pool_A[::-1]+pool_B[::-1]+pool_C[::-1]
        return B
    if user=='c':
        C=pool_A[::-1]+pool_C[::-1]+pool_B[::-1]
        return C

def Filling_Pool(Arr):
    pool_A=Arr[0::3]
    pool_B=Arr[1::3]
    pool_C=Arr[2::3]
    return pool_A,pool_B,pool_C
    

import numpy as np
import random
import time
Random_Array=[]
while len(Random_Array)<=26:
    arr=np.random.randint(1,99)
    if arr not in Random_Array:
        Random_Array.append(arr)
print()
print("Remember any number from it :   ")
print(random.sample(Random_Array,len(Random_Array)))
print()
print()
time.sleep(6)
print()
pool_A,pool_B,pool_C=Filling_Pool(Random_Array)
display(pool_A,pool_B,pool_C)
command_1=input().lower()

First=InWhichPool(pool_A,pool_B,pool_C,command_1)


pool_A1,pool_B1,pool_C1=Filling_Pool(First)
display(pool_A1,pool_B1,pool_C1)
command_2=input().lower()

Second=InWhichPool(pool_A1,pool_B1,pool_C1,command_2)

pool_A2,pool_B2,pool_C2=Filling_Pool(Second)
display(pool_A2,pool_B2,pool_C2)
command_3=input().lower()

Third=InWhichPool(pool_A2,pool_B2,pool_C2,command_3)

magic= (len(Third)-1)/2
time.sleep(2)
print()
print()
print("Hey ! I got you its -->  ",Third[int(magic)])


    
    
```
